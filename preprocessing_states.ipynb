```python
all_states = df.state.unique()
print(len(all_states))
```

```python
def preprocess(lookup, resolution):
  with open(path + 'covid-19-dataset/processed-data/states/processed-us-states-' + lookup + '.csv', 'w+') as lookup_file:

    with open(path + 'covid-19-dataset/processed-data/states/us-states-order.csv', 'w+')  as order_file:

      for state in all_states:
        for date in all_dates:
          filter1 = df[resolution] == state
          filter2 =  df['date'] == date

          if((filter1 & filter2).any()):
            value = df.loc[filter1 & filter2][lookup].to_string(index=False)
          else:
            value = '0'

          if(date == all_dates[-1]):
            print(value , end='')
            lookup_file.write(value)
          else:
            print(value + ', ', end='')
            lookup_file.write(value + ', ')

        if(state=='District of Columbia'):
          state='DC'
        print(state + '\n', end='')
        order_file.write(state + '\n')

        
        print('\n')
        lookup_file.write('\n')
```

```python
########################### careful here 

lookup = 'cases'
resolution = 'state'

preprocess(lookup= lookup, resolution = resolution)
```
