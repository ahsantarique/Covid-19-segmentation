```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid-19-dataset/processed-data/final-us-state-cases.csv')
```

```python
dflog = np.log(df)
```

```python
dflog.fillna(0)
```

```python
from matplotlib.font_manager import FontProperties
from matplotlib.pyplot import figure

df.plot()       
fontP = FontProperties()
fontP.set_size('small')
plt.legend(prop=fontP, loc = 'upper left') 



fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.show()
```

```python
from matplotlib.font_manager import FontProperties
from matplotlib.pyplot import figure

dflog.plot()       
fontP = FontProperties()
fontP.set_size('small')
plt.legend(prop=fontP, loc = 'upper left') 



fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.show()
```
